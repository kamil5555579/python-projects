```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import numpy as np
```

```python
df= pd.read_csv('epa-sea-level.csv', index_col='Year')
```

```python
df
```

```python
fig, ax = plt.subplots(figsize=(12,5), dpi=100)
ax.scatter(x=df.index, y=df['CSIRO Adjusted Sea Level'], color='red', s=1)
ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (inches)')
ax.set_title('Rise in Sea Level')
#df.plot.scatter(x=index, y='CSIRO Adjusted Sea Level')
```

```python
stats=linregress(x=df.index, y=df['CSIRO Adjusted Sea Level'])
```

```python
def f(x):
    return x * stats.slope + stats.intercept

x=np.arange(1880,2050,1)
y=f(x)
```

```python
ax.plot(x,y)
fig
```
