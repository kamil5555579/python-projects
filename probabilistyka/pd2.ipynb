```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funkcja(x):
    return np.exp(-x**2 / 2)

# N oznacza liczbę powtórzeń - liczbę symulacji MC
N = 10**5

# losuję współrzędne $N$ punktów, x w przedziale [-1,1], y w przedziale [0,1]
x_MC = np.random.rand(N)*2 - 1
y_MC = np.random.rand(N)

# sprawdzam które z nich leżą pod wykresem funkcji f(x)
punkty_pod_wykresem = funkcja(x_MC)>y_MC

# liczę pole jako stosunek liczby punktów razy pole prostokąta czyli 2
pole_pod_wykresem = 2 * punkty_pod_wykresem.sum()/N

# wynik podzielony przez pierwiastek z 2 pi
print(f'Pole pod wykresem analitycznie {0.682689 } vs Monte Carlo {pole_pod_wykresem/np.sqrt(2*np.pi):.6f}')
```

```python
fig, ax = plt.subplots(figsize=(9,9), dpi=75)

# robię tablicę x z wartościami [-1,1]
# oraz liczę y dla tych punktów - narysuję w ten sposób funkcję
x = np.arange(-1, 1, 1/1000)
y = funkcja(x)

# rysuję punkty z funkcji
ax.plot(x,y, color='indigo', label=r'funkcja $f(x)=Exp(\frac{-x^2}{2})$')

# rysuję punkty pod wykresem
ax.scatter(x_MC[punkty_pod_wykresem], y_MC[punkty_pod_wykresem], color='red', s=1, label=f'{punkty_pod_wykresem.sum()} punktów pod wykresem')

# upiększanie wykresu
ax.set_xlim(-1,1)
ax.set_ylim(0,1.1)

ax.set_title('Symulacja Monte Carlo')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()

plt.show()
```
